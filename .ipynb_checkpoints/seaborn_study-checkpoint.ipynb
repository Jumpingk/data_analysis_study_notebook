```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sinplot(flip=1): # 
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * 0.5) * (7 - i) * flip) # sin
    plt.show()
sinplot()
```

```python
with sns.axes_style("darkgrid"):  # with 
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
plt.show()
```
